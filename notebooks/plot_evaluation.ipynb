```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import sys
sys.path.append("../scripts")

base_dir = '/Volumes/diedrichsen_data$/data/FunctionalFusion'
if not Path(base_dir).exists():
    base_dir = '/srv/diedrichsen/data/FunctionalFusion'
```

This notebook plots the DCBC and cosine error evaluation of the different parcellations

```python
# Load evaluation results
dcbc = pd.read_csv(base_dir + '/Models/eval2_best.tsv', sep='\t')
coserror = pd.read_csv(base_dir + '/Models/eval_Mdtb.tsv', sep='\t')
```

### DCBC for group arrangement model evaluated on MDTB, Pontine or Nishimoto
Error bars indicate variation in DCBC across subjects

```python

sb.barplot(data=dcbc, x="test_data", y="dcbc", hue="train_data")
# sb.barplot(data=dcbc, x="train_data", y="dcbc", hue="test_data")
```

```python

sb.violinplot(data=coserror, x="test_data", y="coserr_ind2", hue="train_data",
              dodge=True, jitter=True)

coserror.groupby('train_data').coserr_ind2.describe()
```

```python
sb.violinplot(data=coserror, x="test_data", y="coserr_group", hue="train_data",
              dodge=True, jitter=True)

coserror.groupby('train_data').coserr_group.describe()
```
